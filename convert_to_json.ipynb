## Conversion of Wyrmspan Game Cards to JSON format

```python
import pandas as pd
import json
```

Convert xlsx file to json format

```python
cards_df = pd.read_csv('cards.csv')
cards_df
```

```python
# save the dragon cards first in json format
dragon_cards = cards_df[cards_df['type'] == 'Dragon']
dragon_cards = dragon_cards.drop(columns=['type'])

# replace NaNs and change types to int for some columns
for col in ['VP','capacity','Meat','Gold','Crystal','Coin','Egg','Milk']:
    dragon_cards[col] = dragon_cards[col].fillna(0)
    dragon_cards[col] = dragon_cards[col].astype(int)

for col in ['Crimson Cavern','Golden Grotto','Amethyst Abyss']:
    dragon_cards[col] = dragon_cards[col].fillna(False)
    dragon_cards[col] = dragon_cards[col].astype(bool)

dragon_cards = dragon_cards.rename(columns={'ability':'ability_text','Meat':'meat_cost','Gold':'gold_cost','Crystal':'crystal_cost','Coin':'coin_cost','Egg':'egg_cost','Milk':'milk_cost'})

dragon_cards.info()
dragon_cards
```

```python
dragon_dict = dragon_cards.to_dict(orient='records')

for card in dragon_dict:
    ability_string = card['ability_text']
    ability_info = {'adv_effects': {}}
    # split text by commas and periods
    as1 = ability_string.split(',')
    ability_strings = []
    for a in as1:
        ability_strings.extend(a.split('.'))
    # process each string
    for a in ability_strings:
        a = a.lower()
        if 'up to' in a and 'x' in a:
            ability_info['max_uses'] = int(a[-2])
        if 'pay' in a:
            ability_info['cost'] = {}
            ability_info['cost'][a] = 1
        if 'may' in a:
            ability_info['adv_effects']['choice'] = []
        if ('gain' in a or 'lay [egg]' in a or 'tuck' in a) and 'vp' not in a:
            gain_list = ability_info['adv_effects'].get('sequence', [])
            gain_list.append({'gain_': {'type': a, 'amount': 1}})
            ability_info['adv_effects']['sequence'] = gain_list
        if 'opponent' in a:
            ability_info['opponent_effect'] = {'adv_effects': {}}
        if 'if' in a:
            ability_info['condition'] = {'and': [a]}
    # update the card
    if card['abilityType'] == 'Adventurer':
        card['if_activated'] = ability_info
    elif card['abilityType'] == 'OncePerRound':
        card['once_per_round'] = ability_info
    elif card['abilityType'] == 'EndGame':
        card['end_game'] = ability_info
    elif card['abilityType'] == 'WhenPlayed':
        card['when_played'] = ability_info
    card.pop('abilityType')

    if card['size'] == 'Hatchling':
        card['on_feed'] = {'gain_': {'type': '', 'amount': 1}}
        card['on_grow_up'] = {'adv_effects': {'sequence': [{'gain_': {'type': 'vp', 'amount': 1}}]}}
```

```python
with open('dragon_cards_test.json', 'w') as f:
    json.dump(dragon_dict, f, indent=4)
```

Convert cave cards

```python
# save the dragon cards first in json format
cave_cards = cards_df[cards_df['type'] == 'Cave']
cave_cards = cave_cards[['number','ability']]
# rename ability to text
cave_cards = cave_cards.rename(columns={'ability':'text'})

cave_cards.info()
cave_cards
```

```python
cave_dict = cave_cards.to_dict(orient='records')
# we will reformat the ability key to be a list of dicts
# each dict will describe one ability
for card in cave_dict:
    ability_string = card['text']
    ability_info = {'adv_effects': {}}
    # split text by commas and periods
    as1 = ability_string.split(',')
    ability_strings = []
    for a in as1:
        ability_strings.extend(a.split('.'))
    # process each string
    for a in ability_strings:
        a = a.lower()
        if 'up to' in a:
            ability_info['max_uses'] = int(a[-2])
        if 'pay' in a:
            ability_info['cost'] = {}
            ability_info['cost'][a] = 1
        if 'may' in a:
            ability_info['adv_effects']['choice'] = []
        if 'gain' in a or 'lay [egg]' in a:
            gain_list = ability_info['adv_effects'].get('sequence', [])
            gain_list.append({'gain_': {'type': a, 'amount': 1}})
            ability_info['adv_effects']['sequence'] = gain_list
        if 'opponent' in a:
            ability_info['opponent_effect'] = {'adv_effects': {}}
    # update the card
    card['when_played'] = ability_info
```

```python
with open('cave_cards_test.json', 'w') as f:
    json.dump(cave_dict, f, indent=4)
```
