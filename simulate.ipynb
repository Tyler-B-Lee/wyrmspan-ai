```python
import random
import json
```

```python
random.randint(1,4)
```

```python
# read json file
with open('dragon_cards.json') as f:
    d_cards = json.load(f)

with open('cave_cards.json') as f:
    c_cards = json.load(f)
print(c_cards)
```

```python
# simulate drawing a card from a deck
def draw_card(deck):
    card = random.choice(deck)
    deck.remove(card)
    return card
```

```python
d = draw_card(d_cards)
for key, value in d.items():
    print(key, value)
```

```python
c = draw_card(c_cards)
for key, value in c.items():
    print(key, value)
```
